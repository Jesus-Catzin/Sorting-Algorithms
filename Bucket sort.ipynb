<strong><center>Bucket Sort (Worst Case)</center></strong>

<p><strong>The Bucket Sort</strong> algorithm is a sort algorithm that divides the list to order into sub-lists called buckets, which are sorted by another sort algorithm, once ordered these sub-lists are concatenated (put together) to form the ordered list, the pseudocode of the algorithm is presented below</p>

<p>As we can see the original list <strong>A</strong> is divided into n <strong>B<sub><em>i</em></sub></strong> sublists which are then sorted independently one of the other, as can be seen in this case the ordering algorithm is used to sort these sublists is the Insert Sort.</p>

<p>Let us say then that each sublist has <strong>b<sub><em>i</em></sub></strong> length, now well given that the sublists are partitions of <strong>A</strong> and that all together form <strong>A</strong>, we can affirm the following</p>

<table>
<tr>
    <td><strong>n = b<sub>0</sub> + b<sub>1</sub> + b<sub>2</sub> + . . . + b<sub>n2</sub> + b<sub>n-1</sub></strong></td>

<p>Now knowing the above we can say that the worst case would be assumed knowing that all the sublists are

<p>Since each partition is sorted independently, we can treat its complexity individually, so the complexity for each sublist Bi is <strong>O(b<sup>2</sup><sub>i</sub>)</strong>, but we also know the following</p>

## Computational Implementation

```python
def insertionSort(b): 
    for i in range(1, len(b)): 
        up = b[i] 
        j = i - 1
        while j >=0 and b[j] > up: 
            b[j + 1] = b[j] 
            j -= 1
        b[j + 1] = up 
    return b
def bucketSort(x): 
    start = timer()# we start the time 
    arr = [] 
    slot_num = 10 # 10 means 10 slots, each 
# slot's size is 0.1 
    for i in range(slot_num): 
        arr.append([]) 
    # Put array elements in different buckets 
    for j in x: 
        index_b = int(slot_num * j) 
        arr[index_b].append(j) 
    # Sort individual buckets 
    for i in range(slot_num): 
        arr[i] = insertionSort(arr[i]) 
    # concatenate the result 
    k = 0
    for i in range(slot_num): 
        for j in range(len(arr[i])): 
            x[k] = arr[i][j] 
            k += 1
    end = timer() # we stop the time
    time_m = end-start
    return time_m, len(x)
```
