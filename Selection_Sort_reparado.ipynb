```python
from timeit import default_timer as timer
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys
plt.rcParams["figure.figsize"] = (20,10)

# For selection Sort
A = np.random.randint(0,1000, size = 10)

def selectionsort(A):
    t1 = timer()
    # Traverse through all array elements 
    for i in range(len(A)): 

        # Find the minimum element in remaining  
        # unsorted array 
        min_idx = i 
        for j in range(i+1, len(A)): 
            if A[min_idx] > A[j]: 
                min_idx = j 

        # Swap the found minimum element with  
        # the first element         
        A[i], A[min_idx] = A[min_idx], A[i] 
    t2 = timer()
    t3 = t2 - t1
    return t3
```

The algorithm sorts an array by repeatedly finding the minimum element (considering ascending order) from unsorted part and putting it at the beginning. The algorithm maintains two subarrays in a given array.
- The sorted array (left)
- The unsorted array (right).

## Process

Logic: Array is considered into two parts (Unsorted and Sorted)
1. Select the lowest element in the remaining array
2. Bring it tot the first unsorted position
3. Increase the number of position to evaluate by 1

```python
# Sizes: [10,20,30,40,50,100,500,1000,2500,7500,10000,12500,15000]
# Iterate 10 times per size and compute average

# x axis of n size
n_size = np.linspace(start = 2, stop = 1000, num = 100, dtype=int)

# Initialice a list for times 
insertion_times_list = []

for size in n_size: # for each size
    
    # Initialize a temporal list for times per each size
    n_times_list = []
    
    # Iterate 10 times per size
    for i in range(10): 

        arr = np.random.randint(0,10000, size = size)
        #start = timer()
        #selectionsort(arr)
        #end = timer()
        #time = end - start
        n_times_list.append(selectionsort(arr))
        
    # Compute average time and append to the list
    insertion_times_list.append(np.mean(n_times_list))
```

```python
pd.DataFrame(data={'Size': n_size,
                  'Insertion Sort Running Time':insertion_times_list,
                  })
```

```python
plt.plot(n_size, insertion_times_list, label='selection sort')
```
